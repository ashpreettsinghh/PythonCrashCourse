```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = sns.load_dataset('exercise')
```

```python
df.drop('Unnamed: 0',axis=1,inplace=True)
```

```python
sns.distplot(df['pulse'])
```

```python
row,col = df.shape
list1 = []
for i in range(row):
    list1.append(int(df['time'][i].split()[0]))
```

```python
df['int time'] = list1
```

```python
df.drop('time',axis=1)
```

```python
sns.jointplot(x='int time',y='pulse',data=df)
```

```python
sns.pairplot(df,hue='kind',palette='coolwarm',height=5)
```

```python
sns.rugplot(x='int time',y='pulse',data=df,hue='diet')
```

```python
sns.kdeplot(x='int time',y='pulse',data=df,hue='diet')
```

```python
sns.barplot(x='int time',y='pulse',data=df,hue='diet')
```

```python
sns.countplot(x='kind',data=df)
```

```python
sns.boxplot(x='int time',y='pulse',data=df, hue='diet')
```

```python
sns.violinplot(x='int time',y='pulse',data=df,hue='diet',split=True)
```

```python
sns.stripplot(x='int time',y='pulse',data=df)
```

```python
sns.violinplot(x='int time',y='pulse',data=df,hue='diet',split=True,palette='magma')
sns.swarmplot(x='int time',y='pulse',data=df,palette='coolwarm',hue='diet')
```

```python
sns.heatmap(df.corr(),annot=True,lw=3,linecolor='grey')
```

```python
sns.clustermap(df.corr())
```

```python
g = sns.PairGrid(df)
g.map_diag(sns.histplot)
g.map_lower(sns.scatterplot)
g.map_upper(sns.kdeplot)
```
