```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
path = "/Users/ashpreetsingh/Downloads/Diamonds Prices2022.csv"
#Source -- Kaggle data set : https://www.kaggle.com/datasets/nancyalaswad90/diamonds-prices
```

```python
df = pd.read_csv(path)
df.set_index('Unnamed: 0')
```

```python
caratlist = df['carat']
pricelist = df['price']
cutlist = df['cut']
clarity = df['clarity']
depth = df['depth']
```

```python
fig1 = plt.figure()
axes1 = fig1.add_axes([0.5,.5,2,2])
axes1.plot(pricelist,caratlist)
axes1.set_xlabel('Price')
axes1.set_ylabel('Carat')
axes1.set_title('Carat V/S Price')
```

```python
fig,axes = plt.subplots(nrows=3,ncols=1,figsize=(25,25),dpi=100)
axes[0].plot(pricelist,cutlist,color='purple')
axes[0].set_xlabel('Price')
axes[0].set_ylabel('Cuts')
axes[1].plot(pricelist,clarity,lw=10,color='r',ls='--')
axes[1].set_xlabel('Price')
axes[1].set_ylabel('Clarity')
axes[2].plot(pricelist,depth,color='green')
axes[2].set_xlabel('Price')
axes[2].set_ylabel('Depth')

plt.tight_layout()
```

```python
depthl = df['depth']
nfig = plt.figure()
ax1 = nfig.add_axes([1,1,2,1])
ax1.plot(pricelist,caratlist,label='B')
ax1.plot(pricelist,depthl,label='A')
nfig.legend(loc=5)
```
